Lab for today:
choose your favorite 3 book categories from this website: https://books.toscrape.com/
and create a dataframe for each of the 3 categories, storing:
the title of the book
the price of the book
the number of stars it has

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# df1: Fiction 65 - results 1 to 20
```

```python
url_fiction = "https://books.toscrape.com/catalogue/category/books/fiction_10/index.html"
response_1 = requests.get(url_fiction)
response_1
```

```python
soup_fiction = BeautifulSoup(response_1.content)
```

```python
soup_fiction
```

```python
h3_list = soup_fiction.find_all("h3")
```

```python
h3_list
```

```python
title_list = [i.text for i in h3_list]
title_list
```

```python
p_list = soup_fiction.find_all("p", class_="star-rating")
```

```python
p_list
```

```python
rating_list = []

for i in p_list:
    # Añadir la totalidad de la clase del elemento a la lista
    # elemento.get('class') devuelve una lista de las clases, las unimos en una cadena
    clases_all = " ".join(i.get('class'))
    rating_list.append(clases_all)
```

```python
rating_list
```

```python
p_price_list = soup_fiction.find_all("p", class_="price_color")
```

```python
p_price_list
```

```python
price_list = [i.text for i in p_price_list]
```

```python
price_list
```

```python
fiction_dict ={"title": title_list , "rating": rating_list , "price" : price_list}


fiction_df = pd.DataFrame(fiction_dict)
```

```python
fiction_df
```
