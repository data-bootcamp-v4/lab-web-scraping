```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url_fantosy = "https://books.toscrape.com/catalogue/category/books/fantasy_19/index.html"
response = requests.get(url_fantosy)
response
```

```python
soup_fantasy = BeautifulSoup(response.content)
```

```python
#Fantasy

star_converter = {
    "Zero": 0,
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

book_titles = []
book_prices = []
book_stars = []

all_title = soup_fantasy.find_all('ol')
for ol in all_title:
    for li in ol.find_all("li"):
        book_titles.append(li.h3.a.get("title"))
        book_prices.append(li.find("p", class_="price_color").text)
        book_stars.append(
            star_converter.get(
                li.find("p", class_="star-rating").get("class")[1]
            )
        )
```

```python
fantasy_dict = {"title": book_titles, "price": book_prices, "rating": book_stars}
fantasy_df = pd.DataFrame(fantasy_dict)
fantasy_df
```
